```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('telco_customer.csv')
df
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges']
df.info()
```

```python
#Mengubah data menjadi float

df['TotalCharges'] = df['TotalCharges'].astype(float)
df.head()
```

```python
df.info()
```

```python
# Menentukan mean data

mean = df.groupby('InternetService')['TotalCharges'].mean().reset_index(name='Mean')
mean
```

```python
# Menentukan median data

median = df.groupby('InternetService')['TotalCharges'].median().reset_index(name='Median')
median
```

```python
#Menggabungkan data mean dan median serta tidak menampilkan kolom yang duplikat

df_group = pd.concat([mean, median], axis=1)
df_group = df_group.loc[:,~df_group.columns.duplicated()]
df_group
```

```python
dfmelt = df_group.melt(id_vars='InternetService')
dfmelt
```

```python
plt.plot
```

```python
import matplotlib.patches

plt.figure(figsize=(8,8))
splot = sns.barplot(data=dfmelt, x='InternetService', y='value', hue='variable', palette='rainbow')

plt.title('Total Charges based on Internet Services')

for p in splot.patches:
    splot.annotate(format(p.get_height(),'.1f'),
                  (p.get_x() + p.get_width() / 2., p.get_height()),
                  ha = 'center', va = 'center',
                  size=12,
                  xytext = (0,10),
                  textcoords = 'offset points')
    
plt.xlabel('Internet Services', fontsize= 12)
plt.ylabel('Total Charges', fontsize= 12)
```

```python
fig, axes = plt.subplots(1,2, figsize=(14,7), sharey=True)

fig.suptitle('Total Charges based on Internet Services')

splotmean = sns.barplot(ax=axes[0], x = 'InternetService', y='Mean', data=mean, color='red', label='Mean')
axes[0].legend()

#Membuat barplot pada data Mean
for p in splotmean.patches:
    splotmean.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   size=12,
                   xytext = (0, 9), 
                   textcoords = 'offset points')
    
plt.xlabel('Internet Services', fontsize= 12)
plt.ylabel('Total Charges', fontsize= 12)

#Membuat barplot pada data Median
splotmed = sns.barplot(ax=axes[1], x = 'InternetService', y='Median', data=median, color='blue', label='Median')
axes[1].legend()


for p in splotmed.patches:
    splotmed.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   size=12,
                   xytext = (0, 9), 
                   textcoords = 'offset points')
    
plt.xlabel('Internet Services', fontsize= 12)
plt.ylabel('Total Charges', fontsize= 12)
```
